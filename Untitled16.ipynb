```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


csv_filename = "/content/Housing.csv"
df = pd.read_csv(csv_filename)


df["area"] = pd.to_numeric(df["area"], errors="coerce")
df["price"] = pd.to_numeric(df["price"], errors="coerce")


df = df.dropna(subset=["area", "price"])


house_size = df["area"].values.reshape(-1, 1)
house_price = df["price"].values


model = LinearRegression()
model.fit(house_size, house_price)


new_size = np.array([[1200]])
predicted_price = model.predict(new_size)


print("Predicted Price for 1300 sq. ft house: $", round(predicted_price[0], 2))


plt.scatter(house_size, house_price, color="green", label="Actual Data")
plt.plot(house_size, model.predict(house_size), color="blue", label="Regression Line")
plt.xlabel("House Size (sq. ft)")
plt.ylabel("House Price ($)")
plt.legend()
plt.title("Linear Regression: House Size vs Price")
plt.show()
```
